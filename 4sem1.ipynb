```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt

## generate the data and plot it for an ideal normal curve

## x-axis for the plot
x_data = np.arange(-5, 5, 0.001)

## y-axis as the gaussian
y_data = stats.norm.pdf(x_data, 0, 1)

## plot data
plt.plot(x_data, y_data, color="red")
plt.show()
```

```python
from scipy.stats import lognorm
s = 0.68 # задаем форму логнормального распределения
rv = lognorm(s)
r = rv.rvs(size=1000) # генерируем 1000 значений
plt.hist(r, density=True,color='r')
plt.show()
```

```python
distribution = np.linspace(0, np.minimum(rv.dist.b, 2))
#print("Distribution : \n", distribution)
plot = plt.plot(distribution, rv.ppf(distribution),color='r')
```
